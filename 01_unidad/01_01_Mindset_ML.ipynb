# El "Mindset" del ML

## Programación tradicional:
Se establecen las reglas y junto con unos inputs obtenemos el resultado, tal es el caso de la formula lineal para convertir grados centigrados en fahrenheit, una simple función:

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit
```

```python
celsius_to_fahrenheit(100)
```

Ahora bien, supongamos que no conocemos la formula para convertir grados celsius  a fahrenheit, pero si tenemos un dataset con dos columnas: grado centigrado y su equivalencia en grados fahrenheit, usemos nuestra función *`celsius_to_fahrenheit`* para crear este dataset.

```python
# Creemos primero el intervalo de numeros celcius
import numpy as np

celcius = np.arange(1, 1001, 1)
```

```python
# Creamos un diccionario para almacenar las equivalencias
dic = {}

for celsius_value in celcius:
    dic[celsius_value] = celsius_to_fahrenheit(celsius_value)
```

```python
# Obtenemos nuestra tabla
import pandas as pd

df = pd.DataFrame(list(dic.items()), columns=["Celsius", "Fahrenheit"])

df.head(10)
```

## Machine Learning:

Desde ahora nos olvidaremos que existe la formula para convertir grados celcius a fahrenheit, ¿será posible que un algotimo de machine learning tan básico como regresión lineal nos ayude a encontrar los valores que componen la formula?

$y = mx + b$

Donde:

𝑚 = pendiente

𝑏 = intercepto

```python
X = df[["Celsius"]]   # independiente
y = df["Fahrenheit"]  # dependiente
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print("Pendiente (m):", pendiente)
print("Intercepto (b):", intercepto)
```

```python
modelo.predict(pd.DataFrame({"Celsius": [1]}))
```

Como podemos ver el mdoelo encontro esas reglas para el cálculo de grados celcius en fahrenheit.

Este ejercicio demuestra cómo una relación matemática conocida puede ser reconstruida a partir de datos utilizando un modelo de regresión lineal.

A partir de un conjunto de datos generado con valores de temperatura en Celsius y su conversión a Fahrenheit, se entrenó un modelo que logró identificar con precisión la estructura de la relación entre ambas variables. El modelo estimó correctamente una pendiente de aproximadamente 1.7999 y un intercepto cercano a 32.00, lo que coincide con la fórmula original de conversión.
