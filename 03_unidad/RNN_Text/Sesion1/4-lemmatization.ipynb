Lemmatization va más allá del stemmin al hacer un análisis morfológico de las palabras y encontrar raices comunes entre diferentes palabras no al usar la palabra común más larga sino la raíz semántica real de la palabra. For ejemplo, el *lemma* de la palabra *was* (en inglés, pasado de ser o estar) es precisamente *ser* ya que es el mismo verbo pero en diferente tiempo, lo cual es mucho más útil que simplemente trocear las palabras.

Analicemos la siguiente oración:

```python
import spacy

nlp = spacy.load("en_core_web_lg")
doc = nlp("I am a runner running in a race because I love to run since I ran today")
```

```python
for token in doc:
    print(f"{token.text:{20}}{token.pos_:{20}}{token.lemma:<{20}}\t{token.lemma_:{30}}")
```

Observamos que para cada token obtenemos el POS (Part of Speech) y su correspondiente lemma. Nótese que los verbos son correctamente lemmatizados a su raíz.

Practica

```python
# Con Stemming (NLTK)
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer('english')

words = ['running', 'ran', 'easily', 'was', 'better']
for word in words:
    print(f"{word} → {stemmer.stem(word)}")
```

```python
# Con Lemmatization (SpaCy)
doc = nlp("running ran easily was better")
for token in doc:
    print(f"{token.text} → {token.lemma_}")
```

```python
# "running" como verbo
doc1 = nlp("I am running")
print(doc1[2].lemma_)  # → "run"

# "running" como sustantivo
doc2 = nlp("Running is fun")
print(doc2[0].lemma_)  # → "running" (sustantivo, no cambia)
```

Ejemplo 1: Verbos irregulares


```python
doc = nlp("I was going to the store. I went yesterday. I will go tomorrow.")
for token in doc:
    if token.pos_ == "VERB":
        print(f"{token.text:10} → {token.lemma_}")
```

Ejemplo 2: Sustantivos plurales


```python
doc = nlp("The cats are playing. One cat is sleeping.")
for token in doc:
    if token.pos_ == "NOUN":
        print(f"{token.text:10} → {token.lemma_}")
```

Ejemplo 3: Adjetivos comparativos


```python
doc = nlp("This is better than that. The best option is this one.")
for token in doc:
    if token.pos_ == "ADJ":
        print(f"{token.text:10} → {token.lemma_}")
```

Compara stemming vs lemmatization

```python
# Compara stemming vs lemmatization
from nltk.stem.snowball import SnowballStemmer
import spacy

stemmer = SnowballStemmer('english')
nlp = spacy.load('en_core_web_sm')

texto = "I was running easily and feeling better"
doc = nlp(texto)

print("Palabra        | Stemming      | Lemmatization")
print("-" * 50)
for token in doc:
    if token.is_alpha and not token.is_stop:
        stem = stemmer.stem(token.text)
        lemma = token.lemma_
        print(f"{token.text:15} | {stem:15} | {lemma}")
```
